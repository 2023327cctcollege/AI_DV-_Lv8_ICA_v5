```python
# Question 3, Scenario 1

variables = ['python_1', 'python_2', 'ai_1', 'ai_2', 'web', 'database', 'systems']
values = ['ciara', 'peter', 'juan', 'jim', 'jane', 'mary', 'bruce', 'anita']

funds = 4

def consistent(v, a):
    match v:
        case 'python_1':
            if a[v] == 'ciara': return True
        case 'python_2':
            if a[v] in ['peter', 'jane', 'bruce']: return True        
        case 'ai_1':
            if a[v] in ['peter', 'juan', 'jim', 'anita'] and a[v] != a.get('ai_2', None): return True    
        case 'ai_2':
            if a[v] in ['peter', 'juan', 'jim', 'anita'] and a[v] != a.get('ai_1', None): return True    
        case 'web':
            if a[v] in ['juan', 'mary', 'anita']: return True    
        case 'database':
            if a[v] == 'jane': return True    
        case 'systems':
            if a[v] in ['jim', 'mary', 'bruce']: return True

def recursive_backtracking(assignment, solutions):
    if len(assignment) == len(variables):
        return solutions, assignment
    
    first = next(v for v in variables if v not in assignment)
    
    for value in values:
        local_assignment = assignment.copy()
        local_assignment[first] = value

        if consistent(first, local_assignment):
            _, result = recursive_backtracking(local_assignment, solutions)

            if result is not None and len(set(result.values())) <= funds:
                solutions.append(result)
            
    return solutions, None

solutions_3, _ = recursive_backtracking({}, [])

solutions_3
```

```python
# Question 6, Scenarios 1 and 2

scenarios = {
    'Scenario 1': {
        'roles': {
            'Python Programmer': 2,
            'AI Engineers': 2,
            'Web Designer': 1,
            'Database Admin': 1,
            'Systems Engineer': 1,
        }, 
        'candidates': {
            'Peter': ['Python', 'AI'],
            'Juan': ['Web', 'AI'],
            'Jim': ['AI', 'Systems'],
            'Jane': ['Python', 'Database'],
            'Mary': ['Web', 'Systems'],
            'Bruce': ['Systems', 'Python'],
            'Anita': ['Web', 'AI'],
            'Ciara': ['Python'],
        },
        'partners': ['Ciara'],
        'funds': 4
    },
    'Scenario 2': {
        'roles': {
            'Python Programmer': 2,
            'AI Engineers': 3,
            'Web Designer': 1,
            'Database Admin': 1,
            'Systems Engineer': 1,
        }, 
        'candidates': {
            'Peter': ['Python', 'AI'],
            'Juan': ['Web', 'AI'],
            'Jim': ['AI', 'Systems'],
            'Jane': ['Python', 'Database'],
            'Mary': ['Web', 'Systems'],
            'Bruce': ['Systems', 'Python'],
            'Anita': ['Web', 'AI'],
            'Ciara': ['Python'],
        },
        'partners': ['Ciara', 'Juan'],
        'funds': 6
    },
    'active': {
        'roles': {}, 
        'candidates': {},
        'partners': [],
        'funds': 0,
        'name': None
    }
}

saved_solutions = None

root = tk.Tk()

w = 700
h = 500
x = root.winfo_screenwidth()//2 - w//2
y = root.winfo_screenheight()//2 - h//2

root.geometry(f'{w}x{h}+{x}+{y}')
root.resizable(False, False)
root.title('CCT \u2b0c CSP')
root.grid_columnconfigure(0, weight=1)
root.grid_rowconfigure(1, weight=1)

frame_top = tk.Frame(root, height=50)
frame_mid = tk.Frame(root, height=50)
frame_mid_left = tk.Frame(frame_mid, height=50)
frame_mid_left_top = tk.Frame(frame_mid_left, height=50)
frame_mid_left_mid = tk.Frame(frame_mid_left, height=50)
frame_mid_left_bot = tk.Frame(frame_mid_left, height=50)
frame_mid_right = tk.Frame(frame_mid, height=50)
frame_mid_right_top = tk.Frame(frame_mid_right, height=50)
frame_mid_right_mid = tk.Frame(frame_mid_right, height=50)
frame_mid_right_bot = tk.Frame(frame_mid_right, height=50)
frame_bot = tk.Frame(root, height=50)

frame_top.grid_columnconfigure(0, weight=1)
frame_mid.grid_columnconfigure([0, 1], weight=1)
frame_mid.grid_rowconfigure(0, weight=1)
frame_mid_left.grid_columnconfigure(0, weight=1)
frame_mid_left.grid_rowconfigure(1, weight=1)
frame_mid_left_top.grid_columnconfigure(0, weight=1)
frame_mid_left_mid.grid_columnconfigure(0, weight=1)
frame_mid_left_bot.grid_columnconfigure(0, weight=1)
frame_mid_right.grid_columnconfigure(0, weight=1)
frame_mid_right.grid_rowconfigure(1, weight=1)
frame_mid_right_top.grid_columnconfigure(0, weight=1)
frame_mid_right_mid.grid_columnconfigure(0, weight=1)
frame_mid_right_bot.grid_columnconfigure(0, weight=1)
frame_bot.grid_columnconfigure([0, 1], weight=1)

frame_top.grid(row=0, sticky='nesw')
frame_mid.grid(row=1, sticky='nesw')
frame_mid_left.grid(row=0, column=0, sticky='nesw')
frame_mid_left_top.grid(row=0, sticky='nesw')
frame_mid_left_mid.grid(row=1, sticky='nesw')
frame_mid_left_bot.grid(row=2, sticky='nesw')
frame_mid_right.grid(row=0, column=1, sticky='nesw')
frame_mid_right_top.grid(row=0, sticky='nesw')
frame_mid_right_mid.grid(row=1, sticky='nesw')
frame_mid_right_bot.grid(row=2, sticky='nesw')
frame_bot.grid(row=2, sticky='nesw')

tk.Label(frame_top, text="Developed by Mateus Campos\nStudent ID 2023327", font=("Arial", 12), width=w, pady=8).grid(row=0)
tk.Label(frame_mid_left_top, text="ROLES", font=("Arial", 12), width=w//2).grid(row=0, column=0)
tk.Label(frame_mid_right_top, text="CANDIDATES", font=("Arial", 12), width=w//2).grid(row=0, column=0)

def pop_up(flag, *args):
    w = root.winfo_width()//2
    h = root.winfo_height()//2
    x = root.winfo_x() + w//2
    y = root.winfo_y() + h//2

    def close(waiter, top):
        waiter.set(3)
        top.destroy()

    top = tk.Toplevel(root)
    top.geometry(f'{w}x{h}+{x}+{y}')
    top.resizable(False, False)
    top.protocol('WM_DELETE_WINDOW', lambda: close(waiter, top))
    top.grab_set()
    top.grid_columnconfigure(0, weight=1)
    top.grid_rowconfigure([0, 1], weight=5)
    top.grid_rowconfigure(2, weight=1)

    top_top = tk.Frame(top)
    top_mid = tk.Frame(top)
    top_bot = tk.Frame(top)

    top_top.grid(row=0, column=0, sticky='nesw')
    top_mid.grid(row=1, column=0, sticky='nesw')
    top_bot.grid(row=2, column=0, sticky='nesw')

    top_top.grid_columnconfigure(0, weight=1)
    top_mid.grid_columnconfigure([0, 1], weight=1)
    top_bot.grid_columnconfigure([0, 1], weight=1)

    top_top.grid_rowconfigure(0, weight=1)
    top_mid.grid_rowconfigure(1, weight=1)
    top_bot.grid_rowconfigure(0, weight=1)

    waiter = tk.IntVar(top, value=0)

    text = tk.Label(top_top, text="TBC", font=("Arial", 12), wraplength=w*.75)

    label_1 = tk.Label(top_mid, text="TBC", font=("Arial", 12))
    label_2 = tk.Label(top_mid, text="TBC", font=("Arial", 12))
    entry_1 = tk.Entry(top_mid)
    entry_2 = tk.Entry(top_mid)

    btn_ok = tk.Button(top_bot, text='OK', command=lambda: waiter.set(1), width=w//2, bg='green')
    btn_cancel = tk.Button(top_bot, text='CANCEL', command=lambda: close(waiter, top), width=w//2, bg='yellow')

    text.grid(row=0, column=0, sticky='nesw')

    label_1.grid(row=0, column=0, sticky='en')
    entry_1.grid(row=0, column=1, sticky='wn')
    label_2.grid(row=1, column=0, sticky='en')
    entry_2.grid(row=1, column=1, sticky='wn')
    
    btn_ok.grid(row=0, column=0, sticky='nesw')
    btn_cancel.grid(row=0, column=1, sticky='nesw')

    entry_1.focus()
    
    match flag:
        case 0:
            top.title('Add Role')
            text.configure(text='Add a new role to the scenario')
            label_1.configure(text='Title: ')
            label_2.configure(text='Need: ')
            
            btn_ok.wait_variable(waiter)

            if waiter.get() == 1:
                role = entry_1.get().split(';')
                need = entry_2.get().split(';')

                return [{'name': role[i], 'need': need[i]} for i, _ in enumerate(role)], close(waiter, top)
            
            return 'canceled', ''
        case 1:
            top.title('Edit Role')
            text.configure(text='Edit this role in the scenario')
            label_1.configure(text='Title: ')
            label_2.configure(text='Need: ')

            top_bot.grid_columnconfigure([0, 1, 2], weight=1)
            btn_ok.configure(width=w//3)
            btn_cancel.configure(width=w//3)
            btn_delete = tk.Button(top_bot, text='DELETE', command=lambda: waiter.set(2), width=w//3, bg='red')
            btn_delete.grid(row=0, column=2, sticky='nesw')

            entry_1.insert(0, args[0]['name'])
            entry_2.insert(0, args[0]['need'])

            btn_ok.wait_variable(waiter)

            match waiter.get():
                case 1:
                    role = {'name': entry_1.get(), 'need': entry_2.get()}
                case 2:
                    role = None
                case 3:
                    return 'canceled', ''

            return role, close(waiter, top)            
        case 2:
            top.title('Add Candidate')
            text.configure(text='Add a new candidate to the scenario')
            label_1.configure(text='Name: ')
            label_2.configure(text='Skills: ')

            top_mid.grid_rowconfigure(1, weight=0)
            top_mid.grid_rowconfigure(2, weight=1)
            partner = tk.BooleanVar(top)
            label_3 = tk.Label(top_mid, text="Partner: ", font=("Arial", 12))
            label_3.grid(row=2, column=0, sticky='en')
            check = tk.Checkbutton(top_mid, variable=partner)
            check.grid(row=2, column=1, sticky='wn')
            
            btn_ok.wait_variable(waiter)

            if waiter.get() == 1:
                candidate = entry_1.get().split(';')
                skills = entry_2.get().split(';')
                partner = partner.get()

                return [{'name': candidate[i], 'skills': skills[i].split(','), 'partner': partner} for i, _ in enumerate(candidate)], close(waiter, top)
            
            return 'canceled', ''
        case 3:
            top.title('Edit Candidate')
            text.configure(text='Edit this candidate in the scenario')
            label_1.configure(text='Name: ')
            label_2.configure(text='Skills: ')

            top_mid.grid_rowconfigure(1, weight=0)
            top_mid.grid_rowconfigure(2, weight=1)
            partner = tk.BooleanVar(top_mid, value=args[0]['partner'])
            label_3 = tk.Label(top_mid, text="Partner: ", font=("Arial", 12))
            label_3.grid(row=2, column=0, sticky='en')
            check = tk.Checkbutton(top_mid, variable=partner)
            check.grid(row=2, column=1, sticky='wn')
            top_bot.grid_columnconfigure([0, 1, 2], weight=1)
            btn_ok.configure(width=w//3)
            btn_cancel.configure(width=w//3)
            btn_delete = tk.Button(top_bot, text='DELETE', command=lambda: waiter.set(2), width=w//3, bg='red')
            btn_delete.grid(row=0, column=2, sticky='nesw')

            entry_1.insert(0, args[0]['name'])
            entry_2.insert(0, ','.join(args[0]['skills']))

            btn_ok.wait_variable(waiter)

            match waiter.get():
                case 1:
                    candidate = {'name': entry_1.get(), 'skills': entry_2.get().split(','), 'partner': partner.get()}
                case 2:
                    candidate = None
                case 3:
                    return 'canceled', ''

            return candidate, close(waiter, top)
        case 4:
            top.title('Load Scenario')
            text.configure(text='Load a default scenario')
            label_1.destroy()
            label_2.destroy()

            combobox = ttk.Combobox(top_mid)
            combobox['values'] = tuple(scenarios.keys())[:-1]
            combobox.grid(row=0, column=1, sticky='wn')
            combobox.current(0)

            entry_1.destroy()
            entry_2.destroy()

            btn_ok.wait_variable(waiter)

            if waiter.get() == 1:
                return combobox.get(), close(waiter, top)
            
            return 'canceled', ''
        case 5:
            top.title('Solve Scenario')
            text.configure(text='Solve the currently active scenario')
            label_1.configure(text='Funds: ')
            label_2.configure(text='Method: ')

            combobox = ttk.Combobox(top_mid)
            combobox['values'] = tuple(['Recursive Backtracking', 'CSP Framework'])
            combobox.grid(row=1, column=1, sticky='nw')
            combobox.current(0)

            entry_1.insert(0, args[0])
            entry_2.destroy()
            
            btn_ok.wait_variable(waiter)

            if waiter.get() == 1:
                return entry_1.get(), combobox.current(), close(waiter, top)
            
            return 'canceled', '', ''
        case 6:
            label_1.destroy()
            label_2.destroy()
            entry_1.destroy()
            entry_2.destroy()

            if args[0] == None:
                top.title('No Solutions saved')
                text.configure(text='No solutions saved\n\nSolve a scenario first')
                text.grid(row=0, column=0, columnspan=2, sticky='esw')
                btn_cancel.destroy()
            elif len(args[0]) == 0:
                top.title('No Solutions Found')
                text.configure(text='No solutions found\n\nThis scenario cannot be solved')
                text.grid(row=0, column=0, columnspan=2, sticky='esw')
                btn_cancel.destroy()
            else:
                top.title('Solutions')
                text.configure(text=f'{len(args[0])} solution(s) found in {round(args[1], 2)} seconds\n\nChoose a solution to inspect')

                combobox = ttk.Combobox(top_mid)
                combobox['values'] = tuple([f'Solution {i + 1}' for i in range(len(args[0]))])
                combobox.grid(row=0, column=1, sticky='wn')
                combobox.current(0)

                btn_ok.wait_variable(waiter)

                if waiter.get() == 1:
                    return combobox.current(), args[0][combobox.current()], close(waiter, top)
                
                return 'canceled', '', ''

            btn_ok.wait_variable(waiter)
            
            return 'canceled', '', close(waiter, top)
        case 7:
            label_1.destroy()
            label_2.destroy()
            entry_1.destroy()
            entry_2.destroy()
            btn_ok.destroy()

            top.title('Solution Progress')
            text.configure(text='Solution is in progress\n\nThis can take a while')

            pb = ttk.Progressbar(top_mid, orient='horizontal', length=w//2, mode='indeterminate')
            pb.grid(row=0, column=0, columnspan=2)
            pb.start()

            time_start = time.time()

            event = threading.Event()
            threading.Thread(target=get_solutions, args=[args[0], args[1], event]).start()

            progress_check(event, pb, waiter)

            pb.wait_variable(waiter)

            time_end = time.time()

            if waiter.get() == 1:
                return 'completed', time_end - time_start, close(waiter, top)
            
            return 'canceled', '', ''
    
def add_role(role=None):
    if not role:
        roles, _ = pop_up(0)
        if roles == 'canceled': return
        for role in roles: add_role(role)
    else:
        if len(scenarios['active']['roles']) == 0: clear('left', False)
        
        scenarios['active']['roles'].update({role['name']: int(role['need'])})

        btn = tk.Button(
            frame_mid_left_mid,
            text=f"{role['name']}: {role['need']}",
            bg='palegreen',
            font=("Poppins", 11, 'bold')
        )
        
        btn.grid(row=len(scenarios['active']['roles']), sticky='nesw')
        btn.configure(command=lambda btn=btn, role=role: edit_role(btn, role))

def edit_role(btn, role):
    role_up, _ = pop_up(1, role)

    if role_up == 'canceled': return

    if role_up:
        scenarios['active']['roles'].pop(role['name'])

        role = role_up

        scenarios['active']['roles'].update({role['name']: int(role['need'])})

        btn.configure(
            text=f"{role['name']}: {role['need']}",
            command=lambda btn=btn, role=role: edit_role(btn, role)
        )
    else:
        delete_role(btn, role)

def delete_role(btn, role):
    scenarios['active']['roles'].pop(role['name'])
    
    if len(scenarios['active']['roles']) == 0:
        clear('left', True)
    else:
        btn.destroy()

def add_candidate(candidate=None):
    if not candidate:
        candidates, _ = pop_up(2)
        if candidates == 'canceled': return
        for candidate in candidates: add_candidate(candidate)
    else:
        if len(scenarios['active']['candidates']) == 0: clear('right', False)
        
        scenarios['active']['candidates'].update({candidate['name']: candidate['skills']})
        
        star = ''
        
        if candidate['partner']:
            scenarios['active']['partners'].append(candidate['name'])
            star = '\u2605 '

        btn = tk.Button(
            frame_mid_right_mid,
            text=f"{star}{candidate['name']}: {', '.join(candidate['skills'])}",
            bg='palegreen',
            font=("Poppins", 11, 'bold')
        )
        
        btn.grid(row=len(scenarios['active']['candidates']), sticky='nesw')
        btn.configure(command=lambda btn=btn, candidate=candidate: edit_candidate(btn, candidate))

def edit_candidate(btn, candidate):
    candidate_up, _ = pop_up(3, candidate)

    if candidate_up == 'canceled': return

    if candidate_up:
        scenarios['active']['candidates'].pop(candidate['name'])
        if candidate['name'] in scenarios['active']['partners']:
            scenarios['active']['partners'].remove(candidate['name'])
            
        star = ''
        candidate = candidate_up

        scenarios['active']['candidates'].update({candidate['name']: candidate['skills']})
        if candidate['partner'] and candidate['name'] not in scenarios['active']['partners']:
            scenarios['active']['partners'].append(candidate['name'])
            star = '\u2605 '

        btn.configure(
            text=f"{star}{candidate['name']}: {', '.join(candidate['skills'])}",
            command=lambda btn=btn, candidate=candidate: edit_candidate(btn, candidate)
        )
    else:
        delete_candidate(btn, candidate)

def delete_candidate(btn, candidate):
    scenarios['active']['candidates'].pop(candidate['name'])
    
    if candidate['name'] in scenarios['active']['partners']:
        scenarios['active']['partners'].remove(candidate['name'])

    if len(scenarios['active']['candidates']) == 0:
        clear('right', True)
    else:
        btn.destroy()

def load(scenario=None):
    if not scenario:
        scenario, _ = pop_up(4)
        if scenario == 'canceled': return
        scenarios['active']['name'] = scenario
        load(scenarios[scenario])
    else:
        clear(side='both', empty=False)

        for role in scenario['roles']:
            add_role({
                'name': role,
                'need': scenario['roles'][role]
            })
        
        for candidate in scenario['candidates']:
            add_candidate({
                'name': candidate,
                'skills': scenario['candidates'][candidate],
                'partner': candidate in scenario['partners']
            })

        scenarios['active']['funds'] = int(scenario['funds'])

def solve(scenario=scenarios['active']):
    global saved_solutions

    funds, method, _ = pop_up(5, scenario['funds'])
  
    if funds == 'canceled': return

    scenario['funds'] = int(funds)

    variables = [f'{role.split()[0]}_{i + 1}' for role, need in scenario['roles'].items() for i in range(need)]
    values = [candidate for candidate in scenario['candidates']]
    constraints = [lambda *args: True if len(set(args)) <= scenario['funds'] else False]

    if all([variables, values, constraints]):
        aux_1 = copy.deepcopy(variables)
        aux_2 = {}

        for p in scenario['partners']:
            for s in scenario['candidates'][p]:
                ins = [i for i, e in enumerate(aux_1) if e and e.startswith(s)]
                if ins:
                    i = ins[0]
                    constraints.append(lambda *args, p=p, i=i: True if args[i] == p else False)
                    aux_1[i] = None

        for i, s in enumerate(aux_1):
            if s: 
                constraints.append(
                    lambda *args, i=i, s=s: 
                        True if args[i] in [
                            c for c in scenario['candidates'] if c not in scenario['partners'] and s.split('_')[0] in scenario['candidates'][c]
                        ] else False
                )

                for j in aux_1:
                    if j and s.split('_')[0] in j:
                        try:
                            aux_2[s.split('_')[0]].add(i)
                        except:
                            aux_2[s.split('_')[0]] = {i}

        for i in aux_2:
            if len(aux_2[i]) > 1:
                m = list(aux_2[i])[0]
                n = list(aux_2[i])[1] + 1
                constraints.append(lambda *args, m=m, n=n: True if len(args[m:n]) == len(set(args[m:n])) else False)

        if method == 0:
            progress, elapsed, _ = recursive_backtracking_setup(variables, values, scenario['funds'], method)
        else:
            progress, elapsed, _ = csp_framework(variables, values, constraints, method)
    else:
        progress = 'completed'
        elapsed = None
        saved_solutions = []
    if progress == 'completed':
        show_solution(saved_solutions, elapsed)

def csp_framework(variables, values, constraints, method):
    problem = csp.Problem()
    problem.addVariables(variables, values)

    for constraint in constraints:
        problem.addConstraint(constraint, variables=variables)

    return pop_up(7, problem, method)

def recursive_backtracking_setup(variables, values, funds, method):
    problem = {'vars': variables, 'vals': values, 'funds': funds, 'scenario': scenarios['active']['name']}
    return pop_up(7, problem, method)

def consistent(v, a, s):
    if s == 'Scenario 1':
        match v:
            case 'Python_1':
                if a[v] == 'Ciara': return True
            case 'Python_2':
                if a[v] in ['Peter', 'Jane', 'Bruce']: return True        
            case 'AI_1':
                if a[v] in ['Peter', 'Juan', 'Jim', 'Anita'] and a[v] != a.get('AI_2', None): return True    
            case 'AI_2':
                if a[v] in ['Peter', 'Juan', 'Jim', 'Anita'] and a[v] != a.get('AI_1', None): return True    
            case 'Web_1':
                if a[v] in ['Juan', 'Mary', 'Anita']: return True    
            case 'Database_1':
                if a[v] == 'Jane': return True    
            case 'Systems_1':
                if a[v] in ['Jim', 'Mary', 'Bruce']: return True
    else:
        match v:
            case 'Python_1':
                if a[v] == 'Ciara': return True
            case 'Python_2':
                if a[v] in ['Peter', 'Jane', 'Bruce']: return True        
            case 'AI_1':
                if a[v] == 'Juan': return True    
            case 'AI_2':
                if a[v] in ['Peter', 'Jim', 'Anita'] and a[v] != a.get('AI_3', None): return True    
            case 'AI_3':
                if a[v] in ['Peter', 'Jim', 'Anita'] and a[v] != a.get('AI_2', None): return True    
            case 'Web_1':
                if a[v] == 'Juan': return True   
            case 'Database_1':
                if a[v] == 'Jane': return True    
            case 'Systems_1':
                if a[v] in ['Jim', 'Mary', 'Bruce']: return True

def recursive_backtracking(assignment, solutions, problem):
    if len(assignment) == len(problem['vars']):
        return solutions, assignment
    
    first = next(v for v in problem['vars'] if v not in assignment)
    
    for value in problem['vals']:
        local_assignment = assignment.copy()
        local_assignment[first] = value

        if consistent(first, local_assignment, problem['scenario']):
            _, result = recursive_backtracking(local_assignment, solutions, problem)

            if result is not None and len(set(result.values())) <= problem['funds']:
                solutions.append(result)
            
    return solutions, None

def get_solutions(problem, method, event):
    global saved_solutions
    
    if method == 0:
        saved_solutions, _ = recursive_backtracking({}, [], problem)
    else:
        saved_solutions = problem.getSolutions()
    
    time.sleep(1)
    event.set()

def progress_check(event, pb, waiter):
    if event.is_set():
        waiter.set(1)
        pb.stop()
    else:
        pb.after(1000, lambda: progress_check(event, pb, waiter))

def show_solution(solution, elapsed):
    index, solution, _ = pop_up(6, solution, elapsed)
    if index in [None, 'canceled']: return
    inspect(solution, index)

def inspect(solution, index):
    insp = tk.Toplevel(root)
    insp.geometry(f'{w}x{h}+{x+50}+{y+50}')
    insp.resizable(False, False)
    insp.grab_set()
    insp.title(f'Solution {index + 1}')
    insp.grid_columnconfigure(0, weight=1)
    insp.grid_rowconfigure(1, weight=1)

    insp_frame_top = tk.Frame(insp, height=50)
    insp_frame_mid = tk.Frame(insp, height=50)
    insp_frame_mid_left = tk.Frame(insp_frame_mid, height=50)
    insp_frame_mid_left_top = tk.Frame(insp_frame_mid_left, height=50)
    insp_frame_mid_left_bot = tk.Frame(insp_frame_mid_left, height=50)
    insp_frame_mid_right = tk.Frame(insp_frame_mid, height=50)
    insp_frame_mid_right_top = tk.Frame(insp_frame_mid_right, height=50)
    insp_frame_mid_right_bot = tk.Frame(insp_frame_mid_right, height=50)
    insp_frame_bot = tk.Frame(insp, height=50)

    insp_frame_top.grid_columnconfigure(0, weight=1)
    insp_frame_mid.grid_columnconfigure([0, 1], weight=1)
    insp_frame_mid.grid_rowconfigure(0, weight=1)
    insp_frame_mid_left.grid_columnconfigure(0, weight=1)
    insp_frame_mid_left.grid_rowconfigure(1, weight=1)
    insp_frame_mid_left_top.grid_columnconfigure(0, weight=1)
    insp_frame_mid_left_bot.grid_columnconfigure(0, weight=1)
    insp_frame_mid_right.grid_columnconfigure(0, weight=1)
    insp_frame_mid_right.grid_rowconfigure(1, weight=1)
    insp_frame_mid_right_top.grid_columnconfigure(0, weight=1)
    insp_frame_mid_right_bot.grid_columnconfigure(0, weight=1)
    insp_frame_bot.grid_columnconfigure(0, weight=1)

    insp_frame_top.grid(row=0, sticky='nesw')
    insp_frame_mid.grid(row=1, sticky='nesw')
    insp_frame_mid_left.grid(row=0, column=0, sticky='nesw')
    insp_frame_mid_left_top.grid(row=0, sticky='nesw')
    insp_frame_mid_left_bot.grid(row=1, sticky='nesw')
    insp_frame_mid_right.grid(row=0, column=1, sticky='nesw')
    insp_frame_mid_right_top.grid(row=0, sticky='nesw')
    insp_frame_mid_right_bot.grid(row=1, sticky='nesw')
    insp_frame_bot.grid(row=2, sticky='nesw')

    tk.Label(insp_frame_top, text=f'Number of candidates to be hired: {len(set(solution.values()))}', font=("Arial", 12), width=w, pady=8).grid(row=0)
    tk.Label(insp_frame_mid_left_top, text='ROLES', font=("Arial", 12), width=w//2).grid(row=0, column=0)
    tk.Label(insp_frame_mid_right_top, text="CANDIDATES", font=("Arial", 12), width=w//2).grid(row=0, column=0)
        
    i = -1
    for r in scenarios['active']['roles']:
        i += 1
        tk.Button(
            insp_frame_mid_left_bot,
            text=f"{r}",
            bg='palegreen',
            font=("Poppins", 11, 'bold')
        ).grid(row=i, column=0, sticky='nesw')

        tk.Button(
            insp_frame_mid_right_bot,
            bg='gray',
            font=("Poppins", 11, 'bold')
        ).grid(row=i, column=0, sticky='nesw')

        for j in range(scenarios['active']['roles'][r]):
            i += 1
            tk.Button(
                insp_frame_mid_left_bot,
                bg='gray',
                font=("Poppins", 11, 'bold')
            ).grid(row=i, column=0, sticky='nesw')

            c = [c for c in set(saved_solutions[0].values()) 
                 if r.split(' ')[0] in scenarios['active']['candidates'][c]][j]
            
            star = '\u2605 ' if c in scenarios['active']['partners'] else ''

            tk.Button(
                insp_frame_mid_right_bot,
                text=f"{star}{c}",
                bg='palegreen',
                font=("Poppins", 11, 'bold')
            ).grid(row=i, column=0, sticky='nesw')

    tk.Button(
        insp_frame_bot,
        text="CLOSE",
        command=insp.destroy,
        bg="green",
        font=("Poppins", 11, 'bold')
    ).grid(row=0, column=0, sticky='nesw')

def clear(side='both', empty=True):
    match side:
        case 'left':
            panels = [frame_mid_left_mid]
            scenarios['active']['roles'] = {}
        case 'right':
            panels = [frame_mid_right_mid]
            scenarios['active']['candidates'] = {}
            scenarios['active']['partners'] = []
        case 'both':
            panels = [frame_mid_left_mid, frame_mid_right_mid]
            scenarios['active']['roles'] = {}
            scenarios['active']['candidates'] = {}
            scenarios['active']['partners'] = []
            scenarios['active']['funds'] = 0

    for panel in panels:
        for child in panel.winfo_children():
            child.destroy()
        if empty:
            tk.Label(
                panel,
                text=   "This list is empty.\n\n" +
                        "Use the + button below to add entries, or LOAD a scenario.",
                font=("Arial", 12),
                width=w//2,
                wraplength=w//2,
                pady=8
            ).grid(row=0, column=0)

clear()

button_add_roles = tk.Button(
    frame_mid_left_bot,
    text="+",
    command=lambda: add_role(),
    bg="gray",
    font=("Poppins", 11, 'bold')
).grid(row=0, column=0, sticky='nesw')

button_add_candidates = tk.Button(
    frame_mid_right_bot,
    text="+",
    command=lambda: add_candidate(),
    bg="gray",
    font=("Poppins", 11, 'bold')
).grid(row=0, column=0, sticky='nesw')

button_load = tk.Button(
    frame_bot,
    text="LOAD",
    command=lambda: load(),
    bg="blue",
    font=("Poppins", 11, 'bold')
).grid(row=0, column=0, sticky='nesw')

button_clear = tk.Button(
    frame_bot,
    text="CLEAR",
    command=lambda: clear(),
    bg="yellow",
    font=("Poppins", 11, 'bold')
).grid(row=0, column=1, sticky='nesw')

button_solve = tk.Button(
    frame_bot,
    text="SOLVE",
    command=lambda: solve(),
    bg="green",
    font=("Poppins", 11, 'bold')
).grid(row=1, column=0, sticky='nesw')

button_exit = tk.Button(
    frame_bot,
    text="EXIT",
    command=root.destroy,
    bg="red",
    font=("Poppins", 11, 'bold')
).grid(row=1, column=1, sticky='nesw')

root.mainloop()
```
